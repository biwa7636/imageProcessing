```python
import sys
PATH_opencv = '/usr/local/Cellar/opencv/2.4.13/lib/python2.7/site-packages/'
sys.path.append(PATH_opencv)
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

# 1. Generate Gaussian Blurred

```python
# image source: https://llambert363.files.wordpress.com/2015/05/book2_kaleidoscope_photo_dotted-wishes-no-7-kaleidoscope-joy-mckenzie.jpg
# will generate h * w images including the original ones

h,w = 2, 3
img = cv2.imread('board.jpg', 0)
sample_size = 11
sigma = 2

images = []
for i in range(1, h*w + 1):
    image = cv2.GaussianBlur(img, (sample_size, sample_size), i * sigma)
    images.append(image)

    plt.imshow(image,'gray')
    plt.title('image{0}'.format(str(i), str(i+1)))
    plt.show()
```

# 2. Calculate the DOG (difference of gaussian)

```python
from collections import Counter

images_d = []

for i in range(len(images)-1):
    d = images[i] - images[i+1]
    images_d.append(d)
    plt.imshow(d,'gray')
    plt.title('image{0} minus image{1}'.format(str(i+1), str(i+2)))
    plt.show()
```

# 3. Calculate the difference of difference of images (Laplacian)

```python
images_l = []
for i in range(len(images_d)-1):
    l = images_d[i] - images_d[i+1]
    images_l.append(l)
    plt.imshow(l,'gray')
    plt.show()
```
